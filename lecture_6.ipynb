### Аналоги встроенных объектов

Допустим, нам необходимо итерироваться по float-числам. Для типа int в std python предусмотрен оператор range()

```python
[i for i in range(0.2, 0.3, 0.1)]
```

К сожалению он не предусматривает итерирование по float, но мы можем сделать свой!

```python
def frange(x,y,step):
    res = x # Почему нельзя использовать x в качестве возвращаемого параметра?
    while res <= y:
        yield res
        res += step
```

```python
list(frange(0,10,0.2)) # Можем наблюдать проблему точности float
```

```python
def frange(x,y,step):
    res = x
    while res <= y:
        yield round(res, 2) # Используем округление, где в качестве второго параметра используем число знаков после запятой
        res += step
```

```python
list(frange(0,10,0.2))
```

А теперь сделаем тоже самое без извращений

```python
np.linspace(0,10,51) # Здесь в качестве третьего параметров мы передаём общее колличество шагов!
```

```python
ls = np.linspace(-np.pi, np.pi, 10000) 
```

Также возможно создавать numpy-массивы из итерируемых объектов

```python
sin_x = np.fromiter((np.sin(x) for x in ls), dtype=np.float32) # Аналог развёртки списка
```

```python
plt.plot(ls, sin_x, '--', color='r')
```

```python
plt.plot(ls, np.exp(ls))
```

<ol>
<li> (О — большое) — верхняя граница, в то время как (Омега — большое) — нижняя граница. Тета требует как (О — большое), так и (Омега — большое), поэтому она является точной оценкой (она должна быть ограничена как сверху, так и снизу). К примеру, алгоритм требующий Ω (n logn) требует не менее n logn времени, но верхняя граница не известна. Алгоритм требующий Θ (n logn) предпочтительнее потому, что он требует не менее n logn (Ω (n logn)) и не более чем n logn (O(n logn)). </li>
<li> f(x)=Θ(g(n)) означает, что f растет так же как и g когда n стремится к бесконечности. Другими словами, скорость роста f(x) асимптотически пропорциональна скорости роста g(n). </li>
<li> f(x)=O(g(n)). Здесь темпы роста не быстрее, чем g (n). O большое является наиболее полезной, поскольку представляет наихудший случай. </li>
</ol>
